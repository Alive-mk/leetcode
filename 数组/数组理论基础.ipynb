数组是存放在连续空间上同类型数据的集合

```python
def test_arr():
    array = [
        [0, 1, 2],
        [3, 4, 5]
    ]
    for row in array:
        for item in row:
            print("Value:", item, "Address:", id(item))
```

```python
def main():
    test_arr()
```

```python
if __name__ == "__main__":
    main()
```

由于整数对象在Python中是不可变的，并且小的整数可能被内部缓存以便重用，所以不同的变量和元素有可能指向内存中相同的地址。

```python
a = 100
b = 100
print(id(a))
print(id(b))
```

对于更大或者更小的整数值，这种缓存机制不适用。例如：

```python
a = 257
b = 257
print(id(a))
print(id(b))
```

至于整数对象在物理内存中的存储方式，它取决于Python解释器的内存管理策略。一般来说，解释器会试图以有效的方式分配内存，但这并不意味着连续的整数值会被连续地存储在内存中。例如，解释器可能会根据当前内存的使用情况，分配离散的内存块给新的整数对象。这样的内存分配细节通常对Python程序员来说是透明的，因为高级语言的目的之一就是抽象掉这些底层细节。
