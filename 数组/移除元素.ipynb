要知道数组的元素在内存地址中是连续的，不能单独删除数组中的某个元素，只能覆盖。

### 暴力解法：双重循环

暴力解法的时间复杂度是O(n^2)

```python
from typing import List
```

```python
class Solution_baoli:
    def removeElement(self, nums: List[int], val: int) -> int:
        i, l = 0, len(nums)
        while i < l:
            if nums[i] == val:
                for j in range(i+1, l):
                    nums[j-1] = nums[j]
                l -= 1
                i -= 1
            i += 1
        return l
```

在 Python 中没有 i-- 这样的递减运算符。Python 并不支持像 C、C++、Java 那样的自增 (++) 和自减 (--) 运算符。在 Python 中，如果你想递减一个变量，你需要使用 -= 1 这样的赋值运算符

### 双指针算法

快指针其实是一直在动的，fast用于寻找新数组中的元素，而slow用于记录新数组中的元素，注意最后一个slow++后已经是数组的长度了

```python
class Solution_shuangzhizhen:
    def removeElement(self, nums: List[int], val: int) -> int:
        fast, slow = 0, 0
        while fast < len(nums):
            if nums[fast] != val:
                nums[slow] = nums[fast]
                slow += 1
            fast += 1
        return slow
```

注意这些实现方法并没有改变元素的相对位置！

时间复杂度：O(n)
空间复杂度：O(1)

```python
nums_1 = [1,2,2,3,4,5,2,7,8,9]
nums_2 = [1,2,2,3,4,5,2,7,8,9]
```

```python
val = 2
```

```python
solution_1 = Solution_baoli()
solution_2 = Solution_shuangzhizhen()
```

```python
result_1 = solution_1.removeElement(nums_1, val)
result_2 = solution_2.removeElement(nums_2, val)
```

```python
print(f'{result_1}\n{result_2}')
```
